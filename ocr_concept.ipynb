```python
from PIL import Image
import pytesseract
import re

pytesseract.pytesseract.tesseract_cmd = "/opt/local/bin/tesseract"

'''Extract text from image zoomed into feeding rate area'''
def extract_text_from_image(image_path):
    image = Image.open(image_path)
    text = pytesseract.image_to_string(image)
    return text

'''Clean text extracted from image to only include feeding rate'''
def parse_feeding_rate(ocr_text):
    # Normalize common OCR mistakes
    cleaned = ocr_text.replace("+", "").replace("I", "1")

    # Look for a number (2–4 digits is typical for ml/hr)
    match = re.search(r"\b(\d{2,4})\b", cleaned)

    if match: #case where feeding rate is detected
        rate = match.group(1)
        return f"Feeding Rate: {rate} ml/hr"
    else: #case where no feeding rate was detected
        return "Feeding Rate: Not detected"

if __name__ == "__main__":
    print("Example 1: Image Displays Feeding Rate of 124ml/hr")
    image_path = "/Users/tabathaguebard/Desktop/SP CODE/Images/ex1zoomin.png"
    extracted_text = extract_text_from_image(image_path)

    result = parse_feeding_rate(extracted_text)
    print(result)

    print("Example 2: Pump Currently Priming (Not Feeding)")
    image_path = "/Users/tabathaguebard/Desktop/SP CODE/Images/ex2zoomin.png"
    extracted_text = extract_text_from_image(image_path)

    result = parse_feeding_rate(extracted_text)
    print(result)
```
